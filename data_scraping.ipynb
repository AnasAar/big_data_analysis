```python
chrome_options = Options()
chromeOptions = webdriver.ChromeOptions()
#chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--verbose')
chrome_options.add_experimental_option("prefs", {
        "download.default_directory": "/Users/mac/Desktop/s5/big data/files",
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "safebrowsing_for_trusted_sources_enabled": False,
        "safebrowsing.enabled": False
})
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-software-rasterizer')
chrome_options.add_argument("--disable-notifications")
chrome_options.page_load_strategy = 'eager'
ser = Service(chromedriver.install())
```

```python
driver = webdriver.Chrome(options=chrome_options,service=ser)
driver.get('https://www.marchespublics.gov.ma/index.php?page=entreprise.EntrepriseAdvancedSearch&AvisExtraitPV')
try:
        driver.find_element(by=By.XPATH,value='//*[@id="tag-1"]/span/div').click()
except :
        print('---------------no popup-----------------')
```

```python
urls=[]
for i in range(200):
    print('page number : '+str(3300+i)+'-------------------------------')
    try:
        driver.find_element(by=By.XPATH,value='//*[@id="tag-1"]/span/div').click()
    except :
        print('---------------no popup-----------------')
    content = driver.page_source
    soup = BeautifulSoup(content)
    for e in soup.findAll(attrs="actions"):
        try :

            url=e.find('a')['href']
            if url not in urls:
                urls.append(url)
        except TypeError:
            print('-----------------no url in the list----------------')
    driver.find_element(by=By.XPATH,value='//*[@id="ctl0_CONTENU_PAGE_resultSearch_PagerTop_ctl2"]').click()
    sleep(3)
    print('page number : '+str(3300+i)+' scrpaed '+str(len(urls))+' link')
```

```python
import os
def removeDuplicateFiles():
    file_path = r"/Users/mac/Desktop/s5/big data/files"
    file_list = os.listdir(file_path)
    # print (file_list)
    for file_name in file_list:
        if " (1)" not in file_name:
            if " (2)" not in file_name:
                if " (3)" not in file_name:
                    if " (4)" not in file_name:
                        continue
                    original_file_name = file_name.replace(' (4)', '')
                original_file_name = file_name.replace(' (3)', '')    
            original_file_name = file_name.replace(' (2)', '')
        original_file_name = file_name.replace(' (1)', '')
        if not os.path.exists(os.path.join(file_path, original_file_name)):
            continue  # do not remove files which have no original
        #check if the files got the same size, trick to not remove files with difrent content but the same name
        elif  os.stat(os.path.join(file_path, file_name)).st_size == os.stat(os.path.join(file_path, original_file_name)).st_size: 
            os.remove(os.path.join(file_path, file_name))
```

```python
removeDuplicateFiles()
```
